```python
#importing libraries
import os,re,csv
import pandas as pd
```

```python
#loading data
hi = open("dataset/hindi_data.txt", "r")
bhp = open("dataset/bhojpuri_data.txt", "r")
mth = open("dataset/maithali_data.txt", "r")
mrt = open("dataset/marathi_data.txt", "r")
npl = open("dataset/nepali_data.txt", "r")
sns = open("dataset/sanskrit_data.txt", "r")
```

```python
#reading data in decoded format
hindi_data = hi.read().decode("utf-8-sig")
bhojpuri_data = bhp.read().decode("utf-8-sig")
maithali_data = mth.read().decode("utf-8-sig")
marathi_data = mrt.read().decode("utf-8-sig")
nepali_data = npl.read().decode("utf-8-sig")
sanskrit_data = sns.read().decode("utf-8-sig")
```

```python
#cleaning and preprocessing data
def preprocess(data):
    #remove special character
    data=re.sub('\||`|~|;|{|}|]|\[|=|\^|\$|&|\*|<|>|/|\'|#|@|!|%|-|\?|"|,|\(|\)|\.|:|\+|_', ' ', data)
    #remove roman alphabets, digits
    data = re.sub('[a-zA-Z0-9]+', ' ', data)
    #remove non devanagari character
    data=re.sub(ur'[^\u0900-\u097F]+', ' ', data)
    #remove devnagari digits, viram
    data=re.sub(ur'[\u0964-\u096F]+', ' ', data)
    #convert multiple space into single
    data = re.sub('[ ]+', ' ', data)
    return data
```

```python
#preprocessed data
hindi_data = preprocess(hindi_data)
bhojpuri_data = preprocess(bhojpuri_data)
maithali_data = preprocess(maithali_data)
marathi_data = preprocess(marathi_data)
nepali_data = preprocess(nepali_data)
sanskrit_data = preprocess(sanskrit_data)
```

```python
#creation of dictionary which contains bigram and it's frequency
def createbigramfrequency(data):
    listofbigram = []
    for i in range(len(data)):
        if(len(data[i:i+2].strip())>1):
            listofbigram.append(data[i:i+2])
    unique = set(listofbigram)
    dic = {}
    for i in unique:
        dic[i] = 0
        for bi in listofbigram:
            if i == bi:
                dic[i] = dic[i]+1
    return dic
```

```python
hi_bi=createbigramfrequency(hindi_data)
bhp_bi=createbigramfrequency(bhojpuri_data)
mth_bi=createbigramfrequency(maithali_data)
mrt_bi=createbigramfrequency(marathi_data)
npl_bi=createbigramfrequency(nepali_data)
sns_bi=createbigramfrequency(sanskrit_data)
```

```python
#loading dataset from which hindi tweets is to be extracted
data = pd.read_csv("tweets[semi_processed].csv")
```

```python
#selecting the column text
data = data['text']
```

```python
#each tweet is stored into a particular language list, based on the tweet's detected language
hindi_sentence = []
bhojpuri_sentence = []
maithali_sentence = []
marathi_sentence = []
nepali_sentence = []
sanskrit_sentence = []
for line in data:
    hi_count=0
    bhp_count=0
    mth_count=0
    mrt_count=0
    npl_count=0
    sns_count=0
    for word in line.split(" "):
        word=word.decode('utf-8')
        hi_fr=0
        bhp_fr=0
        mth_fr=0
        mrt_fr=0
        npl_fr=0
        sns_fr=0
        for i in range(len(word)):
            if(len(word[i:i+2].strip())>1):
                if(word[i:i+2] in hi_bi.keys()):
                    hi_fr+= hi_bi[word[i:i+2]]
                if(word[i:i+2] in bhp_bi.keys()):
                    bhp_fr+= bhp_bi[word[i:i+2]]
                if(word[i:i+2] in mth_bi.keys()):
                    mth_fr+= mth_bi[word[i:i+2]]
                if(word[i:i+2] in mrt_bi.keys()):
                    mrt_fr+= mrt_bi[word[i:i+2]]
                if(word[i:i+2] in npl_bi.keys()):
                    npl_fr+= npl_bi[word[i:i+2]]
                if(word[i:i+2] in sns_bi.keys()):
                    sns_fr+= sns_bi[word[i:i+2]]
        list_fr=[hi_fr, bhp_fr, mth_fr, mrt_fr, npl_fr, sns_fr]
        if(hi_fr==max(list_fr)):
            hi_count+=1
        if(bhp_fr==max(list_fr)):
            bhp_count+=1
        if(mth_fr==max(list_fr)):
            mth_count+=1
        if(mrt_fr==max(list_fr)):
            mrt_count+=1
        if(npl_fr==max(list_fr)):
            npl_count+=1
        if(sns_fr==max(list_fr)):
            sns_count+=1
    
    list_count=[hi_count, bhp_count, mth_count, mrt_count, npl_count, sns_count]
    if(hi_count==max(list_count)):
        hindi_sentence.append(line)
    elif(bhp_count==max(list_count)):
        bhojpuri_sentence.append(line)
    elif(mth_count==max(list_count)):
        maithali_sentence.append(line)
    elif(mrt_count==max(list_count)):
        marathi_sentence.append(line)
    elif(npl_count==max(list_count)):
        nepali_sentence.append(line)
    else:
        sanskrit_sentence.append(line)
        
```
